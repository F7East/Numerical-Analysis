#Problem 8

```python
def f(t,x):
  A = np.array([[0,-0.2],[0.8,-0.1]])
  return A.dot(x)
```

```python
x0 = np.array([2,0])
solve = ivp(f, (0,60), x0, max_step = 0.01)
```

```python
t,x,y = solve.t, solve.y[0], solve.y[1]
plt.plot(solve.t, x, label = 'Romeo')
plt.plot(solve.t, y, label = 'Juliet')
plt.legend()
```

#Problem 14

```python
def f(t,v):
  [x,y,z,w] = v
  return np.array([z, w, -x/(x**2+y**2)**1.5, -y/(x**2+y**2)**1.5])
```

```python
x0 = [4,0,0,0.5]
solve = ivp(f, (0,50), x0, max_step = 0.25)
```

```python
plt.figure(figsize=(4,4))
plt.plot(solve.y[0],solve.y[1])
```

```python
def euler(y,f,tmax):
  h = 0.0025
  ys = []
  for i in range(int(tmax/h)):
    y = y + h*f(1,y)
    ys.append(y)
  return np.array(ys).transpose()
```

```python
solve = euler(x0, f, 50)
plt.figure(figsize=(4,4))
plt.plot(solve[0],solve[1])
```

```python
ws = [0.5, 0.6, 0.8, 0.4, 0.2, 0.2]
hs = [0.25, 0.5, 0.5, 0.25, 0.25, 0.05]
tmax = [50, 150, 200, 35, 30, 30]
```

```python
for i in range(len(ws)):
  x0 = [4,0,0,ws[i]]
  solve = ivp(f, (0,tmax[i]), x0, max_step = hs[i])
  plt.figure(figsize=(4,4))
  plt.plot(solve.y[0],solve.y[1])
  plt.title(str(ws[i]))
```
